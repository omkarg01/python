```python
import matplotlib.image as mpimg
from matplotlib.offsetbox import TextArea, DrawingArea, OffsetImage, AnnotationBbox

img = mpimg.imread('ipl_image.png')
fig, ax = plt.subplots()


arr_lena = mpimg.imread('ipl_image.png')

imagebox = OffsetImage(arr_lena, zoom=0.7)

ab = AnnotationBbox(imagebox, (18.5, 20))

ax.add_artist(ab)


plt.figure(figsize=(150,100))
# plt.figure(figsize=(20, 10))
# ax1 = fig.add_axes([0,0,3,3])
# fig.axes([0,0,3,3])
# fig.add_axes([0.1,0.1,0.8,0.8])
ax.scatter(overs_at_wick,[18,72,130,168,173],c='k',s=100)
ax.plot(overs, df_mi.groupby('over')['total_runs'].sum().cumsum(), linewidth=3,c="#004ba0" , label = "MI")

ax.plot(overs, df_kkr.groupby('over')['total_runs'].sum().cumsum(),linewidth=3, label = "KKR",c="#b3a123")
ax.scatter(list(df_kkr.groupby('over')['player_dismissed'].first().dropna().index),[0,49,53,63,87,131,153,160], c='k', label='Wickets',s=100)
ax.legend(fontsize='xx-large',prop={'size': 25} )
ax.set_xticks(overs)
ax.set_xlabel("Over's",fontsize=20)
ax.set_ylabel("Runs's",fontsize=20)


fig1 = plt.figure(1)
fig1.text(0.48, 0.90, "Mumbai Indians", ha="right", va="bottom", fontsize =20,color="#004ba0")
fig1.text(0.5, 0.90, "&", ha="center", va="bottom", fontsize =20)
fig1.text(0.52,0.90,"Kolkata Knight Rider's", ha="left", va="bottom", fontsize =20,color="#b3a123")

fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)

ax.tick_params(axis='both', which='major', labelsize=15)
plt.show()
```
