```python
def fetching_code(code):
    list_index = cdf[cdf['Provisional Cutoff List of CAP Round I For AY 2019-20, Published On :08-08-2019'].str.contains(fr'({code})', na=False)].index
    list_index += 1
    return list_index
```

```python
def matching_branch(code):
    dic_index_branch = {} #index as a key and branch as value
    for each_index in fetching_code(code):
    #     final_dict[each_index]
        string = cdf.iloc[each_index, 7]
        match = re.search(r'(Mechanical Engineering|Information Technology|Computer Engineering)', str(string))
        if match != None:
            dic_index_branch[each_index]=match.group(1)      
    return dic_index_branch
```

```python
def matching_caste(code):
    for each_ind, branch_name in matching_branch(code).items():
        list_index_caste_match = list(cdf.iloc[each_ind+1,:].str.extract(r'(GOBC|LOBC)').dropna().index)
        for i in list_index_caste_match:
            match = re.search(r'(\(.+\))',str(cdf.loc[each_ind+2,i]))
            if match != None:
                new_merge_df.loc[idx[code],idx[branch_name, str(cdf.loc[each_ind+1,i])]] = match.group(1)
```

```python
for index in new_merge_df.index:
    matching_caste(index)
```
