```python
df1 = pd.read_csv('..\\ML\\9_decision_tree\\Exercise\\titanic.csv')
```

```python
df1
```

```python
df1.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
```

```python
df1
```

```python
inputs = df1.drop('Survived',axis='columns')
target = df1.Survived
```

```python
inputs
```

```python
inputs.Sex = inputs.Sex.replace({'male':1,'female':0})
```

```python
inputs.Age = inputs.Age.fillna(inputs.Age.mean())
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(inputs,target,test_size=0.2)
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
